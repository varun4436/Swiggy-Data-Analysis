```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount ('/content/gdrive')
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Swiggy.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns
```

```python
df['Total ratings'] = pd.to_numeric(df['Total ratings'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
total_restaurants = df.shape[0]
print('Total Resturants:', total_restaurants)
```

```python
print('Average Ratings:', df['Avg ratings'].mean())
```

```python
print("Average Price:", df['Price'].mean())
```

```python
Food_count = df['Food type'].value_counts().head(5)
```

```python
city_counts = df['City'].value_counts()
print(city_counts)
```

```python
city_counts.head(10).plot(kind = 'bar', color = 'orange')
plt.title("Top Cities by Restaurants")
plt.show()
```

```python
print(df["Food type"].value_counts())
```

```python
Food_type_counts = df['Food type'].value_counts().head(10)
Food_type_counts.plot(kind='bar', color = 'orange')
plt.title("Top Food Types")
plt.show()
```

```python
sns.histplot(df['Avg ratings'], bins=10, color = 'orange')
plt.title("Distribution of Average Ratings")
plt.show()
```

```python
sns.scatterplot(x='Price', y='Avg ratings', data=df, color = 'orange')
plt.title("Average Ratings vs Price")
plt.show()
```

```python
sns.histplot(df['Delivery time'], bins=10, color = 'orange')
plt.title("Distribution of Delivery time")
plt.show()
```

```python
Area_counts = df['Area'].value_counts().head(100)
Area_counts.plot(kind='bar', color = 'orange', figsize=(18, 8))
plt.title("Top Area by Numbers of Restaurants")
plt.xlabel("Area")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(Food_count, labels= Food_count.index, autopct='%1.1f%%')
plt.title("Top Most 5 Food Types")
plt.show()
```
